```python
count = 0
def count_up():
    count+=1
    return count
```

```python
count_up()
```

```python
# local변수와 global 변수
count = 0
def count_up():
    global count
    count+=1
    return count
```

```python
count_up()
```

```python
def divide(n,s):
    quotient = n//s #목
    reamainder = n%s #나머지
    return quotient, reamainder
```

```python
print(divide(10,3))
print(type(divide(10,3)))
```

```python
# 익명함수 lambda
f = lambda x,y: x+y
print(f(1,2))
```

```python
# 중복제거 set() -> 집합
def unique_bus_stop(bus_list):
    return len(set(bus_list))
```

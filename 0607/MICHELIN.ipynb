```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster

file_path = './Data_file/RB_MICHELIN_SLCTN_RSTRNT_INFO_20211231.csv'
michelin = pd.read_csv(file_path)

# 필요없는 데이터 삭제
## 필요없는 데이터 셋
michelin_to_drop_column=['CTPRVN_JLANG_NM','SIGNGU_JLANG_NM',
                         'CTPRVN_SIMCH_CHNLNG_NM','SIGNGU_SIMCH_CHNLNG_NM',
                         'SIGNGU_TRACH_CHNLNG_NM','CTPRVN_CD','SIGNGU_CD',
                        'HMPG_URL','TEL_NO','REPRSNT_MENU_NM','FOOD_SPND_PRICE','BASE_DE']
### 삭제 구문
for column in michelin_to_drop_column:
    michelin = michelin.drop(column, axis=1)
    
#### NaN이 포함된 행의 데이터 삭제
michelin = michelin.dropna()

# 지도 데이터 불러오기
michelin_map = folium.Map(location = [37.566535, 126.97796919999996], zoom_start = 11)

# JSON 파일 불러오기
import json
with open('./미슐랭_JSON/seoul_muncipalities_geo.json', encoding = 'utf-8') as file:
    geo_seoul_gu = json.loads(file.read())
    file.close()

# 맵에 마크 추가
markerMichelin = MarkerCluster().add_to(michelin_map)

# 데이터 파일의 위도와 경도를 읽어 지도에 표시.(가게명, 주소, 미슐랭 등급)
# 지도상 위치가 너무 많으면 숫자로 표기
# 없으면 진한 보라색으로 위치 표시
for idx, val in michelin.iterrows():
            folium.Marker(location = [val.FCLTY_LA, val.FCLTY_LO], 
                          popup = folium.Popup(val.FCLTY_NM + " " + val.RDNMADR_NM + " " + val.MICHELIN_NM, 
                                               max_width = 400), icon = folium.Icon(color = 'darkpurple')).add_to(markerMichelin)

# # 데이터 조회
# michelin

# 지도 조회 (서울 확대 버전)
michelin_map

# # 마크한 지도
# markerMichelin
```
